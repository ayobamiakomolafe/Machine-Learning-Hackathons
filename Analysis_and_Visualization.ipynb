```
import matplotlib.pyplot as plt
```

```
data='/content/drive/My Drive/Bootcamp_participants.csv'
```

```
import pandas as pd
df=pd.read_csv(data)
df.head()
```

GENDER Distribution


```
#Gender visualization
df_gender=df.value_counts('Gender').reset_index()
df_gender
```

```
df_gender.plot(kind='barh', x='Gender')
```

```
#Gender distribution of male and female in respective community
gen=df.groupby('Community')['Gender'].value_counts(ascending=False)
gen.plot(kind='area', figsize=(20,8))
```
